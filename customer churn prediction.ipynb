```python
## Label Encoder -->Churn,TotalCharges,PaymentMethod ,PaperlessBilling , Contract ,  StreamingMovies ,StreamingTV , TechSupport , DeviceProtection,OnlineBackup ,OnlineSecurity ,InternetService,MultipleLines,PhoneService,Dependents,Partner,gender

## Scaling -->TotalCharges,MonthlyCharges ,tenure
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
col1 = ['Churn','TotalCharges','PaymentMethod' ,'PaperlessBilling' , 'Contract' ,  'StreamingMovies' ,'StreamingTV' , 'TechSupport' , 'DeviceProtection','OnlineBackup' ,'OnlineSecurity' ,'InternetService','MultipleLines','PhoneService','Dependents','Partner','gender']
```

```python
df1['TotalCharges'] = pd.to_numeric(df1['TotalCharges'],errors = 'coerce')
```

```python
df1['TotalCharges'].values
```

```python
df1['TotalCharges'].isnull().sum()
```

```python
sns.boxplot(df1['TotalCharges'])
```

```python
df1['TotalCharges'] = df1['TotalCharges'].fillna(df1['TotalCharges'].mean())
```

```python
col2 = ['MonthlyCharges' , 'TotalCharges' ,'tenure']
```

```python
df1[col2].corr()
```

```python
le = LabelEncoder()
```

```python
df1[col1] = df1[col1].apply(le.fit_transform)
```

```python
df1.head()
```

```python
x = df1.drop(['Churn'],axis=1)
```

```python
y = df1['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
```

```python
x_train[col2] = scaler.fit_transform(x_train[col2])
x_test[col2] = scaler.transform(x_test[col2])
```

```python
x_train.head()
```

```python
# Importing the required libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
rc1 = RandomForestClassifier(n_estimators= 242, max_depth= 17, min_samples_split= 8, min_samples_leaf= 7, bootstrap= True)
```

```python
rc1.fit(x_train , y_train)
```

```python
y_pred2 = rc1.predict(x_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
cm = confusion_matrix(y_test,y_pred2)
print(cm)
sns.heatmap(cm,annot=True,cmap='Blues',yticklabels=['No','Yes'],xticklabels=['No','Yes'])
plt.title('Confusion Matrix\n')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
```

```python
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
y_prob = rc1.predict_proba(x_test)[:, 1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
```

```python
roc_auc = auc(fpr, tpr)
```

```python
plt.figure()
plt.plot(fpr, tpr, label="ROC curve (AUC = %0.2f)" % roc_auc)
plt.plot([0, 1], [0, 1], linestyle='--')  # random line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
